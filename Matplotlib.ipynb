Working with subolot, second plot, etc.

> * lengend(loc= 'lower/center/upper left')

> * box_to_anchor (setting the 0,0 to be archor,(1,1) to be top right cornor) 

```python
x=[1,2,3,4,5]
y=[5,4,6,3,2]
y2=[1,2,5,6,4]
plt.plot(x,y,label='Initial Line')
plt.plot(x,y2,label='Now Line')

plt.xlabel('Xlabel')
plt.ylabel('Value')
plt.title('First \n Legend Graph')

#Becuase we already labeled on top
plt.legend(bbox_to_anchor=(1.3,0.7))
```

## Histogram(PDF and CDF)([Return to Module 1](#cell-Plot-toc))

> *Bar chart is ok to comparing, but pretty poor to show distribution*

```python
import matplotlib.pyplot as plt

test_scores = [97,85,86,77,57,36,68,89,36,57,68]
len(test_scores)
```

### Trick to create a list with same number of elements

```python
#Create a list with the same length as test_scores
x=[x for x in range(len(test_scores))]
plt.bar(x,test_scores)
```

### histogram for PDF


```python
import matplotlib.pyplot as plt

# Sample data
test_scores = [97,85,86,77,57,36,68,89,36,57,68]

# Number of bins
num_bins = 5

# Create the histogram for (PDF)
plt.hist(test_scores, bins=num_bins)
# if you want to add  a typical rwidth, then the graph is not connected together


# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
```

```python
import matplotlib.pyplot as plt

# Sample data
test_scores = [97,85,86,77,57,36,68,89,36,57,68]

# Number of bins
num_bins = 5

# Create the histogram for (CDF)
plt.hist(test_scores, bins=num_bins,cumulative=True)


# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
```

## ScatterPlot([Return to Module 1](#cell-Plot-toc))
> * Show correlation or categories(dicision tree)

```python
# Sample data
test_scores = [97,85,86,77,57,36,68,89,36,57,68]
time_spent = [43,16,47,68,36,68,64,53,42,36,52]

plt.scatter(time_spent,test_scores,label='normal tester')

plt.xlabel('Time spent on test')
plt.ylabel('Test score')
plt.title('Test score vs Time spent')


elite_scores=[85,93,86,97,95,89]
elite_time=[52,43,25,46,42,35]
plt.scatter(elite_time,elite_scores,marker='*',label='elite tester')

#plt.legend() bbox_to_anchor=(1, 0.5)
# add the legend on the right side
plt.legend(bbox_to_anchor=(1.35,0.7))
```

#

### Calculate correlation between two lists 
> * Calculate the correlation of list(using Numpy)

> * Pandas corr() calcualte the correlation between two dataframe 

```python
corr = np.corrcoef(test_scores, time_spent)[0,1]
print(corr)
```

## Stack Plot(trick to do Legend) ([Return to Module 1](#cell-Plot-toc))

*Pie chart shows the status at a time, while a stack plot shows status over time.*

```python
year=[1,2,3,4,5,6,7,8]

# numbers in thousands
taxes=[15,64,25,74,365,24,35,347]
overhead=[1,5,26,35,46,57,36,165]
entertainment=[13,25,35,46,68,57,153,263]

plt.stackplot(year,taxes,overhead,entertainment)

# Issue is that we want to label that, but not sure what to label
```

```python
plt.plot([],[],color='m',label='taxes')
plt.plot([],[],color='c',label='overhead')
plt.plot([],[],color='b',label='entertainment')

plt.stackplot(year,taxes,overhead,entertainment,colors=['m','c','b'])
plt.legend(bbox_to_anchor=(1.35,0.8))

plt.title('Company expenses over years')
plt.xlabel('year')
plt.ylabel('Cost, in thousands')
```

## Pie Charts(Starting line, percentage) ([Return to Module 1](#cell-Plot-toc))
> **Plot in angles, colours,etc.**

## Loading Data from a CSV file ([Return to Module 1](#cell-Plot-toc))

```python
import matplotlib.pyplot as plt
import csv

x=[]
y=[]

with open('/content/example.txt','r') as csvfile:
  # create a csv reader file to read the information
  plots=csv.reader(csvfile, delimiter=':')
  for row in plots:
    # because it is text file, it readed as string
    x.append(int(row[0]))
    y.append(int(row[1]))

plt.plot(x,y,label='Loaded from example')
plt.xlabel('Plot number')
plt.ylabel('Random video')
plt.legend()
plt.title('Awesome Graph')
```
